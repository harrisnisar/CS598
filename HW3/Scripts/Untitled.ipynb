```python
import numpy as np
import math
import matplotlib.pyplot as plt
from matplotlib import style
style.use('fivethirtyeight')
```

```python
m1 = np.array([0,0]).reshape((2,1))
m2 = np.array([4,0]).reshape((2,1))
cov1 = np.array([[1,0],[0,1]])
cov2 = np.array([[1,0],[0,1]])
```

```python
def getDiscriminantParameters(classMean, classCovariance, classProbability):
    invClassCovariance = np.linalg.inv(classCovariance)
    Wa = -0.5 * invClassCovariance
    wb = invClassCovariance @ classMean
    wc = -0.5 * (classMean.T @ invClassCovariance @ classMean) - 0.5 * np.log(np.linalg.det(classCovariance)) + np.log(classProbability)
    return (Wa, wb, wc, invClassCovariance)
```

```python
def evalGaussianDiscriminant(x, discrParams):
    Wa = discrParams[0]
    wb = discrParams[1]
    wc = discrParams[2]
    return (x.T @ Wa *x.T).sum(axis=1) + wb.T @ x + wc
```

```python
def evalGuassianPDF(x, mean, cov, inv_cov):
    delta = x - mean
    return np.sqrt(np.linalg.det((2.0*math.pi)*cov))*np.exp( -0.5*(delta.T @ inv_cov * delta.T).sum(axis=1) )
```

```python
p1 = getDiscriminantParameters(m1, cov1, 0.5)
p2 = getDiscriminantParameters(m2, cov2, 0.5)
```

```python
N   = 500
x1  = np.linspace(-10,10,N)
x2  = np.linspace(-10,10,N)
(X1,X2) = np.meshgrid(x1, x2)
U = np.concatenate((X1.reshape(1,N*N),X2.reshape(1,N*N)),0)
G1 = evalGuassianPDF(U, m1, cov1, p1[3]).reshape(N, N)
G2 = evalGuassianPDF(U, m2, cov2, p2[3]).reshape(N, N)
```

```python
fig = plt.figure(figsize=(8,8))

Z = (evalGaussianDiscriminant(U, p1) - evalGaussianDiscriminant(U, p2)).reshape(N,N)
fig = plt.figure()
ax = fig.add_subplot(111)
h0 = ax.contour(X1,X2,Z,cmap='viridis',levels=[0])

h1 = ax.contour(X1,X2,G1,cmap='viridis')

h2 = ax.contour(X1,X2,G2,cmap='plasma')
ax.axis('equal')
plt.show()
```
